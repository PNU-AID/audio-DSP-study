```python
import warnings
import wave
import os
from scipy.io import wavfile
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np
```

```python
cwd = os.getcwd().replace('\\', '/')
cwd = os.path.dirname(cwd).replace('\\', '/')
print(cwd)
```

```python
url = cwd + '/media/sample.wav'
with warnings.catch_warnings():
    warnings.filterwarnings('ignore', category=wavfile.WavFileWarning)
    fs, data = wavfile.read(url)
ipd.Audio(data, rate=fs)
```

```python
with wave.open(url, 'r') as wav_file:
    print(f"File name: {os.path.basename(url)}")
    print(f"Sample rate: {wav_file.getframerate()} Hz")
    print(f"bit depth: {wav_file.getsampwidth()*8} bits")
    print(f"Frame width: {wav_file.getsampwidth()} bytes/sample")
    time = float(wav_file.getnframes()) / wav_file.getframerate()
    print(f"precise time: {time}")
    minute = int(time // 60)
    seconds = int(time % 60)
    print(f"Duration: {minute}:{seconds}")
    print(f"Channels: {wav_file.getnchannels()}")
    print(f"Number of Frames: {wav_file.getnframes()}")
    print(f"Compression Type: {wav_file.getcomptype()}")
    print(f"Compression Name: {wav_file.getcompname()}")
    file_sz = int(wav_file.getframerate() * wav_file.getsampwidth() * 8 * wav_file.getnchannels() * (wav_file.getnframes() / wav_file.getframerate()))
    print(f"file size: {file_sz / 8 / 2**20:.2f} MB")
```

```python
t = np.arange(wav_file.getnframes())
plt.plot(t, data)
plt.xlabel('Time Sample')
plt.ylabel('Quantization Sample')
plt.title('Sampled & Quantized wav')
```
