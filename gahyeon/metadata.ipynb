```python
import wave
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import wavfile
```

```python
fs, data = wavfile.read('song.wav')

with wave.open('song.wav', 'rb') as wav_file:
    sample_rate = wav_file.getframerate()
    quantization_rate = wav_file.getsampwidth() * 8
    duration = wav_file.getnframes() / float(sample_rate)
    num_channels = wav_file.getnchannels()

print(f"Sample rate: {sample_rate} Hz")
print(f"Quantization rate: {quantization_rate} bits")
print(f"Duration: {duration} seconds")
print(f"Number of channels: {num_channels}")
```

```python
t = np.arange(wav_file.getnframes())
plt.plot(t, data)
plt.xlabel('Time Sample')
plt.ylabel('Quantization Sample')
plt.title('Sampled & Quantized wav')
```
